**Finding Outliers through Boxplot :-**

```python
# Specify the numerical columns you want to check for outliers
numerical_columns = ['popularity', 'duration_minutes', 'mode', 'time_signature']

# Create box plots to visualize outliers
plt.figure(figsize=(20, 6))
for column in numerical_columns:
    sns.boxplot(data=Songs, y=column)
    plt.title(f'Box Plot for {column}')
    plt.show()
```

**Since, there are no outliers in other columns but we have outliers in our duration_minutes column.**

```python
# Calculate the IQR for the 'duration_minutes' column
Q1 = Songs['duration_minutes'].quantile(0.25)
Q3 = Songs['duration_minutes'].quantile(0.75)
IQR = Q3 - Q1

# Define the lower and upper bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Find and count the outliers
outliers = Songs[(Songs['duration_minutes'] < lower_bound) | (Songs['duration_minutes'] > upper_bound)]
num_outliers = len(outliers)

print(f'Number of outliers in duration_minutes: {num_outliers}')
```

**CAPPING METHOD TO TREAT OUTLIERS :-**

```python
# Define the percentile value for capping (e.g., 95th percentile)
percentile_value = 95

# Calculate the percentile value for duration_minutes
duration_percentile = np.percentile(Songs['duration_minutes'], percentile_value)

# Replace outliers with the specified percentile value
Songs['duration_minutes_capped'] = np.where(Songs['duration_minutes'] > duration_percentile,
                                            duration_percentile,
                                            Songs['duration_minutes'])
# Now, the 'duration_minutes_capped' column contains capped values for duration_minutes
```

```python
# Specify the numerical columns you want to check for outliers
numerical_columns = ['duration_minutes_capped']

# Create box plots to visualize outliers
plt.figure(figsize=(10, 6))
for column in numerical_columns:
    sns.boxplot(data=Songs, y=column)
    plt.title(f'Box Plot for {column}')
    plt.show()
```

**From the given boxplot we can see that high amout of outliers have been treated.**

```python
# Calculate the IQR for the 'duration_minutes' column
Q1 = Songs['duration_minutes_capped'].quantile(0.25)
Q3 = Songs['duration_minutes_capped'].quantile(0.75)
IQR = Q3 - Q1

# Define the lower and upper bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Find and count the outliers
outliers = Songs[(Songs['duration_minutes_capped'] < lower_bound) | (Songs['duration_minutes_capped'] > upper_bound)]
num_outliers = len(outliers)

print(f'Number of outliers in duration_minutes: {num_outliers}')
```

**WE ARE GOING TO REDUCE THE track_genre UNIQUE VALUES :-**

```python
# Define a mapping dictionary for merging genres and handling missing values
genre_mapping = {
    'acoustic': 'acoustic',
    'rock': 'rock',
    'electronic': 'electronic',
    'pop': 'pop',
    'hip-hop': 'hip-hop',
    'country': 'country',
    'folk': 'folk',
    'jazz': 'jazz',
    'classical': 'classical',
    'latin': 'latin',
    'reggae': 'reggae',
    'metal': 'metal',
    'punk': 'rock',  # Merging punk with rock
    'blues': 'blues',
    'rap': 'hip-hop',  # Merging rap with hip-hop
    'r-n-b': 'r-n-b',
    'soul': 'r-n-b',  # Merging soul with r-n-b
    'dance': 'electronic',  # Merging dance with electronic
    'world-music': 'world-music',
    'holiday': 'holiday',  # New broader category for holiday-themed music
    # For example, you can add a mapping for missing values like this:
    np.nan: 'other',  # Assign missing values to the 'other' category
}

# Create a new column 'merged_genre' based on the mapping
Songs['merged_genre'] = Songs['track_genre'].map(genre_mapping).fillna('other')

# Check the unique values in the 'merged_genre' column
unique_merged_genres = Songs['merged_genre'].unique()
print("Unique merged genres:", unique_merged_genres)
```

# Marking target and predictor variable :-

```python
# Define the target variable (y)
y = Songs['merged_genre_encoded']

# Select predictor variables (X)
X = Songs[['popularity', 'explicit', 'mode', 'time_signature', 'duration_minutes_capped']]

```

# Splitting the dataset for training and testing :-

```python
from sklearn.model_selection import train_test_split
```

```python
X_train , X_test , y_train , y_test = train_test_split(X,y , test_size=0.3 , random_state=42)     #test size=30 percent , train size=70 percent
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
```

# Random Forest :-

```python
# Initialize and train a Random Forest Classifier model
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = model.predict(X_test)

# Evaluate the model's performance
accuracy = accuracy_score(y_test, y_pred)
classification_rep = classification_report(y_test, y_pred)

print(f"Accuracy: {accuracy:.2f}")
print("Classification Report:\n", classification_rep)
```

# KNN :-

```python
from sklearn.neighbors import KNeighborsClassifier
# Initialize and train a K-Nearest Neighbors (KNN) Classifier model
knn_model = KNeighborsClassifier(n_neighbors=5)
knn_model.fit(X_train, y_train)

# Make predictions on the test set using KNN
y_pred_knn = knn_model.predict(X_test)

# Evaluate the KNN model's performance
accuracy_knn = accuracy_score(y_test, y_pred_knn)
classification_rep_knn = classification_report(y_test, y_pred_knn)

print(f"KNN Accuracy: {accuracy_knn:.2f}")
print("KNN Classification Report:\n", classification_rep_knn)
```

# Top 10 Track Names with Artist recommendation from track_genre :-

```python
def recommend_tracks_by_genre():
    # Accept user input for track_genre
    track_genre = input("Enter a track_genre: ")

    # Filter the dataset for the given track_genre
    genre_df = Songs[Songs['track_genre'] == track_genre]

    if genre_df.empty:
        print(f"No tracks found for the genre: {track_genre}")
        return

    # Sort the filtered DataFrame by popularity (you can use any other relevant metric)
    sorted_genre_df = genre_df.sort_values(by='popularity', ascending=False)

    # Get the top N recommendations
    top_n = min(10, len(sorted_genre_df))  # Limit to 10 or less if there are fewer tracks
    top_recommendations = sorted_genre_df.head(top_n)

    # Extract track_name_artists from the recommendations
    recommendations = top_recommendations['track_name_artists'].tolist()

    print(f"Top {top_n} track_name_artists recommendations for genre: {track_genre}")
    for i, recommendation in enumerate(recommendations, start=1):
        print(f"{i}. {recommendation}")

# Example usage:
recommend_tracks_by_genre()
```
